# Fetching Rows and Column from the dataframe.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data.head()
```

Fetching any column (winner column)

```python
data["winner"]
```

```python
type(data['winner']) # One column and row type is a series.
```

Fetching multiple columns from the dataframe

```python
data[["team1", "team2", "winner"]]
```

```python
type(data[["team1", "team2", "winner"]])
```

Fetching row from the dataframe

For this there is a function called as

`data.iloc[index]`

```python
data.iloc[0]
```

```python
type(data.iloc[0])
```

We can also perform slicing to the iloc

```python
data.iloc[1:3] # Type of this would be a dataframe
```

```python
data.iloc[1:5:2]
```

Fancy Indexing from numpy would also be applicable to it.

```python
data.iloc[[1,10,12]]
```

If we want to fetch particular rows and particular columns we can do it via iloc function

The syntax is kinda like this

`data.iloc[row_description, column_description]` 

```python
data.iloc[:,[4, 5, 10]]
```

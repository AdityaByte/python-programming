```python
def get_city_match_count(city: str) -> int:
    return data[data['city'] == city].shape[0]
```

```python
# Masks are the boolean condition
# Suppose we have to fetch the data where the winner team is Royal Challengers Bangalore and the match that was held in Delhi and also the man of the match is Kohli.
# For that we need to prepare three masks for this although you can directly enforce the condition too but this would be more readable.

mask1 = data['winner'] == 'Royal Challengers Bangalore'
mask2 = data['city'] == 'Delhi'
mask3 = data['player_of_match'] == 'V Kohli'

data[mask1 & mask2 & mask3] # For bitwise or do this |
```

# Data Frequency Count
> In this file we gonna learn about how we can count the frequency of any series of data in pandas.

`value_counts()` function is used to count the frequency of any series of data in pandas.

```python
import pandas as pd
```

```python
data = pd.read_csv("dataset.csv")
```

```python
# value_counts() works on series.
data['winner'].value_counts()
```

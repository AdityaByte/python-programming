# Numpy
> Numpy is a low level library written in C (and fortan) for high level mathematical functions. NumPy cleverly overcomes the problem of running slower algorithms on Python by using multidimensional arrays and functions that operate on arrays. Any algorithm can then be expressed as a function on arrays, allowing the algorithms to be run quicky.

### Applications of Numpy

1. A powerful N-dimensional array object.
2. Sophisticated (broadcasting) functions.
3. Tools for integrating C/C++ and fortan code.
4. Useful linear algebra, Fourier transform and random number capabilities.

### Properties of ND Array
1. Homogenous - Only one datatype can be stored.
2. Only be Numbers (Numbers, Float, Complex)
3. Fixed Item Size.

## Indexing With Boolean Arrays

- Suppose a scenerio in which we have to filter those items in an array which are greater than a particular value in that case the indexing with boolean array works.
- This method is also known as filtering as at the end we are filtering items.

```python
import numpy as np
```

```python
arr = np.random.randint(low=1, high=100, size=20).reshape(4,5) # Creating a random array with low value 1 and high value 100 and size would be 20 then reshaping it to 4 rows and 5 columns

arr
```

```python
# Now creating a boolean array with the conditions of the element which are greater than 50
arr > 50
```

```python
# Now what is indexing with boolean array is actually is we mask the boolean array to the original array and as a result we get only those items which are true.
arr[arr>50]
```

Good part about this technique is scalable.

```python
# I want those items which are greater than 50 and which are odd.
arr[(arr > 50) & (arr % 2 != 0)] # It is bitwise and.
```

```python
# If we have to explicitly set the above conditions index numbers to zero
arr[(arr > 50) & (arr % 2 != 0)] = 0
arr
```

## Broadcasting

```python
a = np.arange(10).reshape(2,5) # Two rows and five columns.
b = np.arange(2).reshape(2,1)
print(a, b, sep="\n")
```

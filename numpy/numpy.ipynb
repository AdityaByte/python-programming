# Numpy
> Numpy is a low level library written in C (and fortan) for high level mathematical functions. NumPy cleverly overcomes the problem of running slower algorithms on Python by using multidimensional arrays and functions that operate on arrays. Any algorithm can then be expressed as a function on arrays, allowing the algorithms to be run quicky.

### Applications of Numpy

1. A powerful N-dimensional array object.
2. Sophisticated (broadcasting) functions.
3. Tools for integrating C/C++ and fortan code.
4. Useful linear algebra, Fourier transform and random number capabilities.

### Properties of ND Array
1. Homogenous - Only one datatype can be stored.
2. Only be Numbers (Numbers, Float, Complex)
3. Fixed Item Size.


## Creating Numpy Arrays
There were six methods by which we can create numpy arrays-
1. np.array() - 1D and 2D
2. np.zeros/ones/empty/random
3. np.arange()
4. np.linspace()
5. Copy()
6. Identity()

```python
# Creating numpy/nd array
# Firstly we need to import the numpy library to our code.
import numpy as np # np is an alias
```

```python
# Firsty way for creating the array in numpy is the np.array() method.
arr1 = np.array([1,2,3,4,5])
print(arr1) # Printing the array.
```

```python
# Printing the type of the array.
type(arr1)
```

```python
# Creating a 2-dimensional array.
arr2 = np.array([
    [1,2,3],
    [4,5,6]
])
arr2
```

```python
# Creating the array via the other functions.
# np.zeros(tuple(row, columns))

arr3 = np.zeros((2,3)) # Kinda similar to initializing.
arr3
```

```python
arr4 = np.ones((3,3))
arr4

# NOTE: The by default the datatype created by the ones and zeros method is float.
```

```python
arr5 = np.identity(5)
arr5
```

```python
arr6 = np.arange(10) # Range generates a list in a given range and arange generates an array in a given range.
arr6
```

```python
 # Similar to the range function.
 # np.arange(from, to, step_count)
 arr7 = np.arange(5,20, 2)
 arr7
```

```python
# If we want the lineary spaced numbers then we can use the np.linspace() method.
# np.linspace(lower_bound, upper_bound, numbers_we_want)

arr8 = np.linspace(2,10, 5)
arr8
```

```python
# copy() function for copying a numpy array.
arr9 = arr8.copy()
arr9
```

## Numpy Array VS Python Lists
Numpy arrays has several advantages over normal lists.
1. Faster
2. Convenient
3. Less Memory

#### Numpy arrays takes less memory than traditional lists.

```python
# Creating a list and creating a numpy array and checking how much memory they are taking.

lista = range(100)
nparr = np.arange(100)
```

```python
# For getting the size we have the getsizeof function inside the sys module.
import sys
```

```python
print(sys.getsizeof(87)*len(lista))
```

```python
print(nparr.itemsize*nparr.size)
```
